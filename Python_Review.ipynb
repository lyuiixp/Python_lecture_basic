````python
# for문 예시
test_repeat = [1,2,3,4]

# 요소 접근 시
print("요소 직접 접근")
for x in test_repeat:
    print(x)

print("----------------------")

# 인덱스로 접근 시
print("range 활용")
for x in range(len(test_repeat)):
    print(test_repeat[x])

print("----------------------")

# 인덱스와 값 동시 접근 시
print("enumerate 활용")
for x,y in enumerate(test_repeat):
    print(x,y)
````

#### **List comprehension**
- 컴프리헨션 문법을 통해 편리하고 직관적으로 표현, 속도에 약간 이점
- 기존 사용법
```python

     new_list.append(변수 표현식)
```


````python
# 조건에 맞는 값을 가지는 새로운 리스트 생성
# 10이상의 값만 가지는 새로운 리스트 만들기
default_list = [1,10,100]

new_list = [x if x>=10 else 'a' for x in default_list]      # 1 ,10, 100
new_list

# if else구문을 사용한다면?
# 100 이상은 그대로, 100 미만은 2를 곱한 값을 가지는 리스트
[x if x>= 100 else x*2 for x in default_list]
````

- 재귀 함수
>- 함수 내에서 자기 자신을 다시 호출해 작업 수행

````python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

factorial(4)        # 4 * 3 * 2 *1 
````

#### Lambda 표현식
- 일회성으로 사용하거나 짧은 코드 구현 시 유용

```python
lambda 인자: 표현식
```

-----------------
- **map()** : 리스트 같은 반복가능 객체 각 요소에 함수 적용

````python
# 기본 람다함수
lambda_square = lambda x: x**2 
lambda_square(3)
````

````python
# 요소의 제곱한 값 리스트
test_map = [5,6,7]
squared_list = list(map(lambda_square, test_map))
squared_list
````

````python
# 요소들 중 홀수만 뽑은 리스트
odd_list = list(filter(lambda i : i%2!=0, test_map))
odd_list
````

````python
# 리스트 내 가장 큰 수 찾기
from functools import reduce

reduce(lambda x,y: x if x>y else y, test_map)
````

## 10. 클래스
- 클래스는 인스턴스 변수와 메소드를 정의하는 틀
- __init__() : 객체가 생성될 때 자동으로 호출되는 생성자. 객체의 속성을 초기화해줌.
- self : 클래스 내에서 정의된 인스턴스 변수, 메소드를 참조할 때 사용.
```python
class Car:
    def __init__(self, color, speed):
        self.color = color      # 속성(인스턴스 변수)
        self.speed = speed      # 속성
    
    def accelerate(self):       # 메소드(함수)
        brake = self.brake()    # self.을 붙여 클래스 내 메소드 참조!
        self.speed += 10

    def brake(self):
        self.speed -= 10

mycar = Car()
mycar.accelerate()
print(mycar.speed)
```

````python
class Car:
    def __init__(self, color, speed):
        self.color = color      # 속성(인스턴스 변수)
        self.speed = speed      # 속성
    
    def accelerate(self):       # 메소드(함수)
        brake = self.brake()    # self.을 붙여 클래스 내 메소드 참조!
        self.speed += 10

    def brake(self):
        self.speed -= 10

mycar = Car('레드',90)
mycar.accelerate()
print(mycar.speed)
````

### **상속**
- 기존의 클래스를 재사용하여 새로운 클래스를 만듬
- 새로운 클래스는 기존 클래스의 속성과 메소드를 물려받아 별도의 정의 없이 사용 가능
- 자식 클래스에서 부모 클래스의 메소드를 재정의(오버라이딩)해서 사용 가능

#### 다형성
- 같은 이름의 메소드가 객체에 따라 다르게 동작

#### 캡슐화
- 속성과 메소드에 외부에서 직접적으로 접근하지 못하게 하는 것
- 언더바(_), 더블언더바(__) 사용
>- _ : 클래스 내부적으로만 사용하겠다를 의미
>- __ : 클래스 외부에서 접근 불가

````python
# 상속 예시
# 부모 클래스
class Animal:
    def __init__(self, name):
        self.name = name

    def sound(self):
        return None             

    def describe(self):
        print(f"I am an animal named {self.name}.")

# 자식 클래스 1: Dog
class Dog(Animal):
    # super() 활용
    def __init__(self, nam, species):
        super().__init__(nam)
        self.species = species

    def sound(self):
        # super().sound()
        return "Woof!"
    
    def describe(self):
        super().describe()                  # "I am an animal named {self.name}."
        print(f'내 종은 {self.species}야')

# 자식 클래스 2: Cat
class Cat(Animal):
    def sound(self):
        return "Meow!"

# 객체 생성 및 메서드 호출
dog = Dog("Moka", '웰시코기')
# cat = Cat("Coco")

# dog.describe()           # 출력: I am an animal named Moka.
print(dog.sound())       # 출력: Woof!

# cat.describe()           # 출력: I am an animal named Coco.
# print(cat.sound())       # 출력: Meow!
````

## 11. 정규표현식
- 특정 패턴을 가진 문자열을 찾거나 치환할 때 사용
- re 모듈
```python
import re

re.match('패턴', '문자열') : 문자열의 시작에서 패턴을 찾는다
re.search('패턴', '문자열') : 문자열 전체에서 패턴을 검색
re.findall('패턴', '문자열') : 패턴과 일치하는 모든 부분을 리스트로 반환
re.sub('패턴', '수정할 문자열', '문자열', 바꿀 횟수) : 패턴과 일치하는 부분을 다른 문자열로 치환
re.split('패턴', '문자열') : 패턴을 기준으로 문자열을 분할
```

````python
import re

# 매치
phone_num = "010-1234-5678"
result_match = re.match(r"\d+-\d+-\d+", phone_num)
print(result_match.group())     # 출력: 010-1234-5678

# 검색
text_search = "전화번호: 010-1234-5678"
result_search = re.search(r"\d{3}-\d{4}-\d{4}", text_search)
print(result_search.group())    # 출력: 010-1234-5678

# 패턴을 찾고 리스트 반환
text_findall = "전화번호: 010-1234-5678"
result_findall = re.findall(r"\d+", text_findall)
print(result_findall)           # 출력: ['010', '1234', '5678']

# 치환
text_sub = "내 전화번호는 010-1234-5678 입니다."
masked = re.sub(r"\d{4}-\d{4}", "****", text_sub)
print(masked)                   # 출력: 내 전화번호는 010-****-**** 입니다.
````
