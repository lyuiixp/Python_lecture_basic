```python
# 예제 1
# 과일가게에 과일 재고 정보가 들어있는 딕셔너리가 주어질 때
# 사용자로부터 과일 이름을 입력받아 그 과일이 재고에 있는지 확인하세요.
# 재고에 있으면 'oo 과일의 재고가 o개 남았습니다' 출력
# 재고가 없다면 'oo 과일의 재고가 없습니다.' 출력

fruits = {"banana" : 10, "orange" : 5, "melon" : 9, "grape" : 3}

fruit = input("과일 이름을 입력하세요: ")
if fruit in fruits:     # True or False
    print(f'{fruit} 과일의 재고가 {fruits[fruit]}개 남았습니다.')
else:
    print(f'{fruit} 과일의 재고가 없습니다.')

print(f'{fruit} 과일의 재고가 {fruits[fruit]}개 남았습니다.') if fruit in fruits else print(f'{fruit} 과일의 재고가 없습니다.')
```

```python
# 예제 2
# 중복된 숫자와 문자들로 이루어진 리스트가 존재할 때 각 숫자를 key로 사용하고 
# 각 숫자들이 count된 값을 value로 사용한 딕셔너리를 출력하세요.
num_list = [1, 2, 2, 2, 4, 4, 3, 6, 5, 5, 3, 1, 1, 'A','A']

count_dict = {}
num_set = set(num_list)
for x in num_set:
    if type(x) == int:
        value = num_list.count(x)
        count_dict[x] = value

count_dict
```

```python
# 예제 3
# 사람의 이름과 나이가 담긴 딕셔너리가 존재할 때, 나이를 기준으로 성인과 미성년자 그룹으로 분류한다.
# 이 때 성인들의 이름이 담긴 리스트 adults, 미성년자들의 이름이 담긴 리스트 kids 를 생성하고 출력하시오.
# 성인의 기준은 19세 이상
people = {"Alice": 15, "Bob": 23, "Charlie": 8, "David": 30, "Eve": 19}
adults = []
kids = []

for k,v in people.items():
    if v>=19:
        adults.append(k)
    else:
        kids.append(k)

adults, kids
```

```python
# 예제 4
# 중복된 숫자들이 포함된 리스트가 존재할 때, 중복이 제거된 리스트를 내림차순으로 정렬하여 출력하고,
# 중복 제거된 리스트의 길이가 5 이상일 경우, '리스트 요소가 너무 많습니다'를 출력하시오.
num_list = [5, 3, 8, 3, 2, 5, 1, 9, 8]

duplicated_list = list(set(num_list))
sorted_list = sorted(duplicated_list, reverse = True)

if len(sorted_list) >= 5:
    print('리스트 요소가 너무 많습니다')
```
