## 1. 비교 연산자
- 비교 연산자의 역할: 두 값을 비교하여 참(True) 또는 거짓(False)의 값을 반환.

- 주요 비교 연산자:

>- ==: 같은지 비교 (값이 같으면 참)
>- !=: 다른지 비교 (값이 다르면 참)
>- <, >: 크고 작음 비교
>- <=, >=: 작거나 같다, 크거나 같다

- **비교 연산의 결과는 항상 True 또는 False로 반환된다**
- ==와 =의 차이점 : ==는 비교 연산자, =는 대입 연산자.

## 2. 논리 연산자 
- 논리 연산자의 역할: 여러 조건을 조합하여 복합적인 조건을 만들 때 사용.

- 주요 논리 연산자:

>- and: 모두 참일 때 결과가 참 (True).
>- or: 하나라도 참이면 결과가 참 (True).
>- not: 참을 거짓으로, 거짓을 참으로 반전.

- **and는 모든 조건이 참이어야 참이고, or는 하나의 조건만 참이어도 참.**
- not은 조건을 반전시키는 연산자로, 조건문에서 중요한 역할을 함.
- 

```python
# 비교 연산( >, = , !)
1 = 1   # 대입연산자
```

```python
# AND(&) 예시
personal = {"이름": '철수', '나이': 20, '키' : 180, '사는지역' : '서울', '성별': '남자'}

if True & personal['사는지역'] == '서울':
    print('소개 받을게요')

# OR(|) 예시
if (personal['나이'] >19) | (personal['키'] <190):
    print(True)
        
# NOT(!) 예시
if personal['사는지역'] != '부산':
    print('소개가능')
```

## 3. 조건문 (If, Else, Elif)
- 조건문의 역할: 프로그램의 흐름을 조건에 따라 다르게 만들 수 있음. 조건이 참일 때 특정 코드를 실행하고, 그렇지 않을 때 다른 코드를 실행.
- 구조:
>- if : **조건이 참**일 때 코드 실행.
>- elif(else if) : **추가 조건**을 확인하고, 참일 때 코드 실행.
>- else : **모든 조건이 거짓**일 때 실행되는 블록.

- 조건문은 들여쓰기가 중요! 
- **elif**는 여러 조건을 순차적으로 평가할 때 사용.
- **else**는 조건이 모두 거짓일 때 처리하는 기본값 역할.

```python
# 실습문제 1
# 두 숫자를 입력받고 각 숫자를 비교하여 출력하시오.

num1 = int(input("첫 번째 숫자를 입력하세요: "))
num2 = int(input("두 번째 숫자를 입력하세요: "))

if num1 > num2:
    print(f"{num1}은 {num2}보다 큽니다.")
elif num1 < num2:
    print(f"{num1}은 {num2}보다 작습니다.")
else:
    print("두 숫자는 같습니다.")
```

```python
# 문제: 사용자로부터 키(미터)와 몸무게(킬로그램)를 입력받아 BMI(Body Mass Index)를 계산하고, 그에 따른 건강 상태를 출력하는 프로그램을 작성하세요.
# BMI 계산 공식: BMI = 몸무게(kg) / (키(m) * 키(m))
# BMI에 따른 상태:
# - 18.5 미만: 저체중
# - 18.5 이상 24.9 이하: 정상
# - 25 이상 29.9 이하: 과체중
# - 30 이상: 비만
height = float(input("키를 입력하세요"))
weight = float(input(" 몸무게를 입력하세요:"))

bmi = weight/(height*height)

if bmi < 18.5:
    print("저체중입니다.")
elif 18.5<= bmi <=24.9:
    print("정상") 
elif 25<=bmi<=29.9:
    print("과체중")
else:
    print("비만")

print(f"BMI 지수는 {bmi}입니다")
```

```python
data = [('a', 1), ('b', 2), ('c', 3), ('a', 4),('a', 5)]

dict_1 ={}

for i in data:
    if i[0] not in dict_1:
        dict_1[i[0]] = i[1]
    else:
        if type(dict_1[i[0]]) == list:
            dict_1[i[0]].append(i[1])
        else:
            dict_1[i[0]] = [dict_1[i[0]], i[1]]

print(dict_1)
```
